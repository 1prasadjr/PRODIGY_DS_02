## **TASK 2**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('test.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["Embarked"],inplace=True)
data['Cabin'].fillna("Unknown",inplace=True)
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
data.isnull()
```

##**PLOT FOR SURVIVAL AGE CATEGORIES**

```python
plt.figure(figsize=(6,3))
sns.histplot(data=data,x="Age",kde=True, color="black")
plt.xlabel("AGE")
plt.ylabel("COUNT")
plt.title("AGE DISTRIBUTION")
plt.show()
```

## **SCATTER PLOT FOR AGE AND FARE**

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=data ,x="Age", y="Fare", hue="Age", color="PURPLE")
plt.xlabel("AGE")
plt.ylabel("FARE")
plt.title("PLOT FOR AGE AND FARE")
plt.show()
```

## SURVIVAL RATIO OF GENDERS

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data, x="Sex", hue="Sex")
plt.xlabel("GENDER")
plt.ylabel("COUNT")
plt.title("GENDER DISTRIBUTION")
plt.legend(title= "Survived",loc="upper right")
plt.show()
```
